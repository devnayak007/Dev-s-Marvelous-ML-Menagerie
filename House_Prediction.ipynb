Importing Dependancies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```



> Importing The Boston House Price Dataset



```python
df1= pd.read_csv('Boston-house-price-data.csv')
df1.head()
```

```python
df1=df1.rename(columns={"MEDV":"Price"})
df1.head()
```

***Checking the number of rows and columns.***

```python
df1.shape
```

***Checking the mising Values***

```python
df1.isnull().sum()
```

***Statistical Measures of the Data***

```python
df1.describe()
```

Understanding the Correlation between various features in the Dataset
1. Positive Correlation
2. Negative Correlation

```python
correlation= df1.corr()
```

```python
#constructing heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

***Splitting Data and Target***

```python
X = df1.drop(['Price'],axis=1)
Y = df1['Price']
```

```python
print(X)
print(Y)
```

***Splitting Data into Training and Testing Data***

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

XGBoost Regressor

```python
#loading the model
model = XGBRegressor()
```

```python
#training the model with X_train
model.fit(X_train,Y_train)
```

***Evaluation***

```python
#accuracy for the prediction on training data
df1_prediction = model.predict(X_train)
```

```python
print(df1_prediction)
```

```python
# R squared Error
score_1 = metrics.r2_score(Y_train, df1_prediction)

#Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, df1_prediction)

print("R Squared Error : ", score_1)
print("Mean Absolute Error: ", score_2)
```

```python
#accuracy for the prediction on test data
df1_test_prediction = model.predict(X_test)
```

```python
# R squared Error
score_1 = metrics.r2_score(Y_test, df1_test_prediction)

#Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_test, df1_test_prediction)

print("R Squared Error : ", score_1)
print("Mean Absolute Error: ", score_2)
```

Visualizing the actual Prices and Predicted prices

```python
plt.scatter(Y_test,df1_test_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
plt.show()
```
