***Importing the Dependancies***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
customer_data= pd.read_csv('/content/hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv')
customer_data.head()
```

```python
customer_data.shape
```

```python
customer_data.info()
```

```python
customer_data.describe()
```

```python
data= customer_data.iloc[:,3:5]
data.head()
```

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(10,7))
plt.title('Dendrogram')
dendrogram=sch.dendrogram(sch.linkage(data,'ward' ))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
labels_=cluster.fit_predict(data)
```

```python
labels_
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data['Annual Income (k$)'],data['Spending Score (1-100)'],c=labels_,cmap='rainbow')
```
